```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
z=np.sin(x)
```

```python
%matplotlib inline
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('X Values')
plt.ylabel('sine and cosine')
plt.legend(['cosine','sine'])
```

```python
plt.figure(2)
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Cosine')
plt.xlabel('X')
plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('Sine')
plt.xlabel('X')
```
