```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20);
print(x)
```

```python
m = 1;
c = 2;
y=m*x + c;

print(y)
```

```python
plt.plot(x,y);
plt.xlim((0,25));
plt.ylim((0,50));
```

```python
m_guess = 0.4;
c_guess = -1;
lr = 0.01; # Learning rate

for x in np.arange(10):
    for i in range(50):
        print("Iteration : %d and x : %d"%(i,x));
        #print("m before : %f"%m_guess);
        #print("b before : %f"%c_guess);

        y_pred = m_guess*x+c_guess;
        y_act = m*x+c;
        dEdC = 2*(y_pred-y_act);
        dEdM = x*dEdC;

        err = ((y_pred-y_act)**2);

        print("error is : %f "%err);
        m_new = m_guess - lr*dEdM;
        c_new = c_guess - lr*dEdC;
        
        y_pred_new = m_new*x+c_new;
        err_new = ((y_pred_new-y_act)**2);
        print("new error is : %f "%err_new);
        
        if err_new < err :
            m_guess = m_new;
            c_guess = c_new;
        else :
            break;
            
        print("m after : %f"%m_guess); 
        print("b after : %f"%c_guess);
```
